1. Write a program that accepts a list from user and print the alternate element of list.

```python
print("Weekly Tasks")

total_tasks = int(input("Enter total number of tasks: "))

tasks_list = []

# Taking student names as input
while len(tasks_list) < total_tasks:
    task_name = input("Task name: ")
    tasks_list.append(task_name)

print("\nAll weekly tasks:", tasks_list)

print("(alternate print):")
todays_tasks = []

# Printing alternate task
todays_tasks.append(tasks_list[0:len(tasks_list):2])


print("Today's Tasks List: ", todays_tasks)
```

2. Write a program that accepts a list from user. Your program should reverse the content of list and
display it. Do not use reverse() method.

```python
max_std = 5
std_list = []

while len(std_list) < max_std:
    std = input("Enter student name: ")
    std_list.append(std)
print(f"FIFO order student queue: {std_list}")

# reverse order

reverse_list = []
while len(std_list) > 0:
    reverse_list.append(std_list.pop())
print(f"LIFO order student queue: {reverse_list}")
```

3. Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard.

```python
print("Finding max height - tallest student")

std_total = int(input("Total no. of Students: "))
std_heights = []


for std in range(0, std_total):
    std_height = int(input(f"Student{std} Height: "))
    std_heights.append(std_height)

# getting max value from the list
print(f"Heights: {std_heights}")
def calculate_max_height(tallest_std = std_heights[0]):
    for height in std_heights:
        if height > tallest_std:
            tallest_std = height
    return tallest_std

print(f"The height of tallest student: {calculate_max_height()} ft.")


```
