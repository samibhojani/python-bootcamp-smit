```python
# Build a calculator
choice = int(input("Select operation: 1-SUM,\n2-SUBTRACT,\n3-MULTIPLY,\n4-DIVIDE"))

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))


if choice == 1:
    num3 = num1 + num2
elif choice == 2:
    num3 = num1 - num2
elif choice == 3:
    num3 = num1 * num2
elif choice == 4:
    num3 = num1 / num2
else: print("Wrong input!")

print(f"Result: {num3}")
```

```python
# Mini Project 2 — Smart Salary Tax Checker

salary = int(input("Enter your salary: "))

if salary < 49999:
    tax = 0
    print(f"Tax is {tax} because salary is less than 50K.")
if salary >= 50000 and salary < 100001:
    tax = salary * 0.10
    print(f"10% tax will be applied on {salary}")
if salary > 100000:
    tax = salary * 0.20
    print(f"20% tax will be applied on {salary}")

print(f"Your salary: {salary},\nYour tax amount: {tax},\nSalary after tax deduction: {salary - tax}")
```

```python
# Create a list of expenses
no_of_expense_entries = int(input("Enter no. of entries: "))
expenses = []

for expense in range(no_of_expense_entries):
    expense = int(input(f"Add expense {expense}: "))
    expenses.append(expense)
    

print(expenses)
```

```python
#Mini Project 5 — Modular Calculator
#Create functions for:
#add
#subtract
#multiply
#divide


def add(num1, num2):
    return num1 + num2

def sub(num1, num2):
    return num1 - num2

def mul(num1, num2):
    return num1 * num2

def divide(num1, num2):
    return num1 / num2


cal1 = int(input("Enter number: "))
cal2 = int(input("Enter number2: "))

addition = add(cal1, cal2)
subtraction = sub(cal1, cal2)
multiplication = mul(cal1, cal2)
division = divide(cal1, cal2)

print(f"Results: \nAddition: {addition}\nSubtraction: {subtraction}\nMultiplications: {multiplication}\nDivision: {division}")
```
