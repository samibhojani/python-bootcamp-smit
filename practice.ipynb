```python
products = {"apple": 205, "mango": 160, "oranges": 350, "banana": 50, "kajoo": 1150, "grapes": 120}

expensive_product_price = 200

expensive_products = {}
normal_price_products = {}

for product, price in products.items():
    if price > expensive_product_price:
        expensive_products[product] = price
    else:
        normal_price_products[product] = price
print("Expensive products: ", expensive_products)
print("Normal Price products: ", normal_price_products)
```

```python
print("Contact Book")
contact_book = {}
while True:
    choice = int(input("1-Add Contact\n2-Search Contact\n3-View All\n4-Quit"))

    if choice == 1:
        contact_name = input("Contact Name: ")
        contact_phone_number = int(input(f"{contact_name}'s Phone Number: "))
        contact_city = input(f"{contact_name}'s City: ")
        contact_book [contact_name] = {"phone": contact_phone_number, "city": contact_city}
    elif choice == 2:
        search_contact = input("Enter name to search in contacts: ")
        if search_contact in contact_book:
            print(f"Search result: Found {search_contact}!")
            print(f"{search_contact}'s Phone and City: {contact_book[search_contact]}")
        else: print(f"Search result: Not Found {search_contact}!")
    elif choice == 3:
        if contact_book == {}:
            print(f"Empty contact book, {len(contact_book)} contacts.")
        else:
            print(f"{len(contact_book)} Contacts Available in your phonebook.")
            for contact, details in contact_book.items():
                print(f"{contact}: {details}")
    elif choice == 4:
        print("Exiting phonebook...")
        break
    else: print("ERROR: Invalid input, type 1,2,3 or 4.")
        
```

```python
# functions - single parameter

student_list = [
    "Ali",
    "Sara",
    "Ahmed",
    "Fatima",
    "Usman",
    "Ayesha",
    "Hamza",
    "Zainab",
    "Bilal",
    "Hira"
    ]

# main function is greet all, and sub-function is greet
def greet_all_students():
    def greet(name):
        print(f"Good morning! {name}")
# for loop to call the greet function for all students.
    for student in student_list:
        greet(student)

greet_all_students()
```

```python
# Mini project with functions with multiparameters:

print("Contact Book")
contact_book = {}

def add_contact(contact_name2, contact_phone_number2, contact_city2): # parameters
    contact_book [contact_name2] = {"phone": contact_phone_number2, "city": contact_city2}

def search(search_contact):

    if search_contact in contact_book:
        print(f"Search result: Found {search_contact}!")
        print(f"{search_contact}'s Phone and City: {contact_book[search_contact]}")
    else: print(f"Search result: Not Found {search_contact}!")

def view_all():
        if contact_book == {}:
            return print(f"Empty contact book, {len(contact_book)} contacts.")
        else:
            print(f"{len(contact_book)} Contacts Available in your phonebook.")
            for contact, details in contact_book.items():
                print(f"{contact}: {details}")
     

while True:
    choice = int(input("1-Add Contact\n2-Search Contact\n3-View All\n4-Quit"))

    if choice == 1:
        add_contact(input("Add Contact Name: "), #argument1
                    int(input("Add Phone Number: ")), #argument2
                    input("Add City: ") #argument3
                    )
    elif choice == 2:
        search(search_contact=input("Enter name to search in contacts: ")) # key argument
    elif choice == 3:
        view_all()
    elif choice == 4:
        print("Exiting phonebook...")
        break
    else: print("ERROR: Invalid input, type 1,2,3 or 4.")
        
```
