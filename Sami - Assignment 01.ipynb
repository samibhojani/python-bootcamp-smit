3. Write a program in which use all the operators we can use in Python

```python
# Write your code here
print("----------Python Operators----------------")
print("----------Employee Salary Calculation-----------")

# Arithmetic operators

basic_salary = 150000
house_rent = 42000
conveyance_allowance = 5000

# addition
total_earnings = basic_salary + house_rent + conveyance_allowance

pf = 15000
eobi = 550
health_ins = 1500

# division & multiplication
itax = 5 / 100 * total_earnings 

total_deduction = pf + eobi + health_ins + itax

# subtraction
net_amount = total_earnings - total_deduction

# power
overseas_salary = net_amount ** 3

# floor division
avg_pak_salary_for_the_role = overseas_salary // 2

# modulus / remainder
total_earnings % total_deduction

print(f"Total Salary: {total_earnings}\nTotal Deductions: {total_deduction}\nNet Salary: {net_amount}\nOverseas Salary for similar role: {overseas_salary}\nAverage salary in Pakistan for same role: {avg_pak_salary_for_the_role}")

# Comparison operators
print("Taxation Details")
if basic_salary > 50000:
    print("Tax Payer")
if itax != 0:
    print("Filer")
print("Health package assignment")
if health_ins < 500:
    print("Non accidental not covered")
elif health_ins == 1000:
    print("Basic Health Package")
elif health_ins >= 1500:
    print("Executive Health Package")
elif health_ins <= 0:
    print("Error in entry")

# logical operators
if basic_salary >= 50000 and itax > 0:
    print("Employee is a tax payer and filer")
if health_ins == 15000 or health_ins > 15000:
    print("Executive health support")
print("------------------------------END---------------------------------------------")
print("I know this program becomes longer, i just received my payslip so used it as an example.")

```

4. Completes the following steps of small task:
    - Mention Marks of English , Islamiat and Maths out of 100 in 3 different variables
    - Mention Variable of Total Marks and assign 300 to it
    - Calculate Percentage

```python
# Write your code here
print("-------Student's Obtained Marks---------")
eng = 90
maths = 86
islamiat = 95
marks_obtained = eng + maths + islamiat
total_marks = 300
print(f"English: {eng} out of 100.\nMaths: {maths} out of 100.\nIslamiyat: {islamiat} out of 100.\nTotal marks obtained: {marks_obtained} out of {total_marks}")

# calculating the percentage
total_percentage = marks_obtained / total_marks * 100
print(f"Student scored: {total_percentage:.1f} %")
```
