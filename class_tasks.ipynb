```python
# Exercise 4 — Contact Book  Mini Project
# Build a contact book using a dictionary + input() + a while loop.
# Features:
# 1 Add a contact (name, phone, city)
# 2 Search for a contact by name
# 3 Show all contacts
# 4 Quit

contact_book = {}
while True:
    choice = int(input("1-ADD CONTACT\n2-SEARCH CONTACT\n3-SHOW ALL CONTACTS\n4-QUIT"))

    if choice == 1:  # add new contact
        contact_name = input("Enter Contact Name: ")
        contact_phone = int(input("Enter Phone Number: "))
        contact_city = input("Enter city: ")
        
        contact_book[contact_name] = {"phone": contact_phone, "city": contact_city}
        print(f"A new Contact has been added to your Contact Book")

    elif choice == 2: # search contact
        search_contact = input("Type contact name to search: ")
        if search_contact in contact_book:
            print("Contact found!\nDetails:")
            print("Name: ", search_contact)
            print("Phone: ", contact_book[search_contact]["phone"],"\nCity: ", contact_book[search_contact]["city"])
        else: print(f"{search_contact} not found in Contacts!")
    elif choice == 3:
        print(f"Contact book total contacts: {len(contact_book)}")
        for name, details in contact_book.items():
                print("Name:", name)
                print("Phone:", details["phone"])
                print("City:", details["city"])
                print("-------------------")
    elif choice == 4:
        print("Exiting Contact Book....")
        break
    else: print("ERROR: Invalid input! enter 1,2,3 or 4.")
    
```

```python
# GET STUDENT GRADES - FUNCTION + LOOP

students = [
    {"name": "Ali", "score": 93},
    {"name": "Ahmed", "score": 82},
    {"name": "Aliya", "score": 74},
    {"name": "Asif", "score": 65},
    {"name": "Azad", "score": 57},
    {"name": "Aniqa", "score": 48},
    {"name": "Akbar", "score": 33},
    {"name": "Abdul", "score": 20},
    {"name": "Anfal", "score": 8},
]


def get_grades(score):
    if score >= 90 and score < 100:
        return "Grade A"
    elif score >= 80 and score < 90:
        return "Grade B"
    elif score >= 70 and score < 80:
        return "Grade C"
    elif score >= 60 and score < 70:
        return "Grade D"
    else: return "Grade F"

get_grades(95)

for student in students:
    grade = get_grades(score = student["score"])
    print(f"Student: {student["name"]}\nMarks: {student["score"]}\nGrade: {grade}\n------------------")
```
